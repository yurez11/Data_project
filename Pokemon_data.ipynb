# 1. Столбчатая диаграмма: Средняя атака по типам покемонов



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Pokemon.csv')
plt.figure(figsize=(12, 6))
sns.barplot(x='Type 1', y='Attack', data=df, ci=None, palette='viridis')
plt.title('Средняя атака по типам покемонов', fontsize=16)
plt.xlabel('Тип покемона', fontsize=14)
plt.ylabel('Средняя атака', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# 2. Линейный график: Изменение средней атаки по поколениям для каждого типа

```python
type_generation_stats = df.groupby(['Type 1', 'Generation'])['Attack'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(x='Generation', y='Attack', hue='Type 1', data=type_generation_stats, marker='o', palette='tab20')
plt.title('Изменение средней атаки по поколениям для каждого типа', fontsize=16)
plt.xlabel('Поколение', fontsize=14)
plt.ylabel('Средняя атака', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(linestyle='--', alpha=0.7)
plt.legend(title='Тип покемона', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

# 3. Круговая диаграмма: Распределение покемонов по типам

```python
plt.figure(figsize=(8, 8))
type_counts = df['Type 1'].value_counts()
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Распределение покемонов по типам', fontsize=16)
plt.tight_layout()
plt.show()
```

# 4. Линейчатая диаграмма: Средние характеристики покемонов по поколениям

```python
generation_stats = df.groupby('Generation')[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']].mean()
plt.figure(figsize=(12, 6))
generation_stats.plot(kind='bar', colormap='Paired', alpha=0.8)
plt.title('Средние характеристики покемонов по поколениям', fontsize=16)
plt.xlabel('Поколение', fontsize=14)
plt.ylabel('Средние значения', fontsize=14)
plt.xticks(rotation=0, fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
